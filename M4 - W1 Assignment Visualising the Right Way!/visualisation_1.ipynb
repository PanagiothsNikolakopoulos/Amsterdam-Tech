```python
import csv
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import datetime

# Set the working directory
os.chdir("C:\\Users\\ManosIeronymakisProb\\OneDrive - Probability\\Bureaublad")

# Read the CSV file into a Pandas DataFrame
filepath = "brexitor2.csv"
df = pd.read_csv(filepath)

# Remove the first three rows (which contain irrelevant information) from the DataFrame
df = df.drop(index=[0,1,2])

# Rename the columns of the DataFrame
df = df.rename(columns={'Question: In hindsight, do you think Britain was right or wrong to vote to leave the EU?': 'date', 'Unnamed: 1': 'right', 'Unnamed: 2': 'wrong'})

# Create two empty lists to hold the values of 'Right' and 'Wrong' responses
right, wrong = [], []

# Open the CSV file and create a CSV reader object to extract the values of 'Right' and 'Wrong'
with open(filepath, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row[1].isdigit():
            right.append(float(row[1]))
        if row[2].isdigit():
            wrong.append(float(row[2]))

# Extract the desired column of dates as a list
with open(filepath, newline='') as csvfile:
    reader = csv.reader(csvfile)
    date_list = [row[0] for row in reader]

# Remove the first four rows (which contain irrelevant information) from the date list
date = date_list[4:]

# Change the format of the dates from 'dd/mm/yy' to 'yyyy-mm-dd'
new_date_list = []
for date_str in date:
    date_obj = datetime.strptime(date_str, '%d/%m/%y')
    new_date_str = date_obj.strftime('%Y-%m-%d')
    new_date_list.append(new_date_str)

# Create a scatter plot with blue dots for right and red dots for wrong
fig, ax1 = plt.subplots()
ax1.plot(new_date_list, right, color='blue', label='Right')
ax1.plot(new_date_list, wrong, color='red', label='Wrong')

# Set the title of the plot
ax1.set_title("In hindsight, do you think Britain was right or wrong to vote to leave the EU?")

# Set the y-axis limits and ticks for the first y-axis
ax1.set_ylim(40, 50)
ax1.set_yticks([40, 42, 44, 46, 48, 50])

# Create a second y-axis on the right side
ax2 = ax1.twinx()

# Set the y-axis limits and ticks for the second y-axis
ax2.set_ylim(40, 50)
ax2.set_yticks([40, 42, 44, 46, 48, 50])

# Remove the left y-axis and move the ticks to the right side
ax1.spines['left'].set_visible(False)
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position('right')

# Add legends for 'Right' and 'Wrong' responses
right_legend = ax1.legend(loc="lower center", bbox_to_anchor=(0.5, 0.0))
wrong_legend = ax1.legend(loc="upper center", bbox_to_anchor=(0.5, 1.0))


# Set the x-axis tick labels
ax1.set_xticklabels(["","","","","","",2016,"","","","","","","","","","","","","","","","","","","","","","","","","","","","",17,"","","","",
                     "","","","","","","","","","","","","","","","","","","","","","","","",18])

# Add light gray horizontal lines at all y values
for y_val in range(40, 51):
    ax1.axhline(y=y_val,  color='gray', linestyle='-', alpha=0.2)

# Show the plot
plt.show()


# Set the x-axis tick labels
ax1.set_xticklabels(["","","","","","",2016,"","","","","","","","","","","","","","","","","","","","","","","","","","","","",17,"","","","",
                     "","","","","","","","","","","","","","","","","","","","","","","","",18])

# Add light gray horizontal lines at all y values
for y_val in range(40, 51):
    ax1.axhline(y=y_val,  color='gray', linestyle='-', alpha=0.2)

# Show the plot
plt.show()




















```
