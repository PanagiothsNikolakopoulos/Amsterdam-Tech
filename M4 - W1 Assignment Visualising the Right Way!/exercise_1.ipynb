```python
"""
import libraries
"""
import numpy as np
import matplotlib.pyplot as plt

def z_func(x_coords: np.ndarray, y_coords: np.ndarray) -> np.ndarray:
    """
    A function that takes in two numpy arrays x_coords and y_coords, and returns a numpy
    array of the same shape as x_coords and y_coords. The function uses the equation
    z = sin(sqrt(x^2 + y^2)) / sqrt(x^2 + y^2) to calculate the values of z.

    Args:
        x_coords (numpy.ndarray): The x-coordinates of the points.
        y_coords (numpy.ndarray): The y-coordinates of the points.

    Returns:
        numpy.ndarray: The values of z calculated using the given equation.
    """
    r_values = np.sqrt(x_coords ** 2 + y_coords ** 2)
    z_values = np.sin(r_values) / r_values
    return z_values

# Create numpy arrays x_coords and y_coords
x_coords = np.linspace(-10, 10, 1000)
y_coords = np.linspace(-10, 10, 1000)

# Calculate the values of z using z_func
z_values = z_func(x_coords, y_coords)

# Create the scatter plot
fig, ax = plt.subplots()
sc = ax.scatter(x_coords, y_coords, c=z_values, cmap='coolwarm')

# Add a colorbar to the plot
cbar = plt.colorbar(sc)

# Set plot title and axis labels
ax.set_title("Advanced Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Save plot as a PNG image with 300 DPI resolution
fig.savefig('advanced_scatter_plot.png', dpi=300)
```
